```python
pattern = PII
```

```python
extracted =[] 
for index, row in df.iterrows():
    tempList =[] 
    rowSplit = row['policy_text'].split(" ")
    for val in rowSplit:
        if val in pattern:
            tempList.append(val)
    if len(tempList)>0:
        extracted.append(' '.join(tempList))
    else:
        extracted.append('N/A')

df['PII_Extracted'] = extracted
```
